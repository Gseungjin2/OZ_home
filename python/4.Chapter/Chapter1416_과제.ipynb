[문제7]

0부터 9까지의 숫자 중 일부가 들어있는 정수 배열 numbers가 주어집니다.   
numbers에서 찾을 수 없는 0부터 9까지의 숫자를 모두 찾아 더한 수를 출력해주세요

numbers = [5,8,1,0,6,9]

```python
numbers = [5,8,1,0,6,9]

full_set = set(range(10))  # 0부터 9까지의 숫자를 포함하는 집합을 생성
numbers_set = set(numbers)  # 주어진 배열을 집합으로 변환
missing_numbers = full_set - numbers_set  # 두 집합의 차집합을 구함
result = sum(missing_numbers)  # 차집합의 모든 원소를 더함

print(result)
```

[문제 1: 붕어빵 재고 확인]

붕어빵 가게의 현재 재고를 확인하는 기능을 구현하세요.

요구사항:
- 현재 붕어빵의 재고를 딕셔너리로 관리합니다.
- 딕셔너리에는 붕어빵의 종류와 각각의 재고 개수가 저장됩니다.
- 현재 재고를 화면에 출력하세요.

붕어빵 재고
- 팥붕어빵 : 10개
- 슈크림붕어빵 : 8개
- 초코붕어빵 : 5개

```python
# 붕어빵 재고 딕셔너리
bungeo_stock = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}

# 현재 붕어빵 재고 출력
print("붕어빵 재고")
for bungeo, stock in bungeo_stock.items():
    print(f"- {bungeo} : {stock}개")
```

[문제 2: 붕어빵 주문 받기]

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
# 주문 받기 함수
def take_order():
    order = {}  # 빈 딕셔너리 생성
    while True:
        bungeo = input("주문할 붕어빵 종류를 입력하세요 (종료: '끝'): ")
        if bungeo == "끝":
            break
        quantity = int(input(f"{bungeo}의 개수를 입력하세요: "))
        order[bungeo] = order.get(bungeo, 0) + quantity

    # 주문 내역 출력
    print("\n주문 내역:")
    for bungeo, qty in order.items():
        print(f"- {bungeo}: {qty}개")

# 주문 함수 호출
take_order()
```

[문제 3: 붕어빵 주문 처리]

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
# 초기 붕어빵 재고 설정
bungeo_stock = {
    '팥붕어빵': 10,
    '슈크림붕어빵': 8,
    '초코붕어빵': 5
}

# 주문 처리 함수
def process_order(order):
    for bungeo, qty in order.items():
        if bungeo in bungeo_stock and bungeo_stock[bungeo] >= qty:
            bungeo_stock[bungeo] -= qty
            print(f"{bungeo} {qty}개 판매 완료")
        elif bungeo in bungeo_stock and bungeo_stock[bungeo] < qty:
            print(f"{bungeo} 재고가 부족하여 {qty}개 주문 불가")
        else:
            print(f"{bungeo}는 판매하지 않는 상품입니다.")

# 주문 예시
order_example = {'팥붕어빵': 3, '슈크림붕어빵': 2, '초코붕어빵': 1}
process_order(order_example)
print("\n업데이트된 붕어빵 재고:", bungeo_stock)
```

[문제 4: 여러 손님 주문 처리]

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
# 초기 붕어빵 재고 설정
bungeo_stock = {
    '팥붕어빵': 10,
    '슈크림붕어빵': 8,
    '초코붕어빵': 5
}

# 주문 처리 함수
def process_order(order):
    for bungeo, qty in order.items():
        if bungeo in bungeo_stock and bungeo_stock[bungeo] >= qty:
            bungeo_stock[bungeo] -= qty
            print(f"{bungeo} {qty}개 판매 완료")
        elif bungeo in bungeo_stock and bungeo_stock[bungeo] < qty:
            print(f"{bungeo} 재고가 부족하여 {qty}개 주문 불가")
        else:
            print(f"{bungeo}는 판매하지 않는 상품입니다.")

# 여러 손님 주문 처리
while True:
    order_input = input("주문을 입력하세요 (ex. '팥붕어빵 3', '종료'): ")
    if order_input == '종료':
        break
    try:
        bungeo, qty = order_input.split()
        qty = int(qty)
        process_order({bungeo: qty})
    except ValueError:
        print("잘못된 입력입니다. 다시 입력해주세요.")

# 현재 재고 출력
print("\n업데이트된 붕어빵 재고:")
for bungeo, stock in bungeo_stock.items():
    print(f"{bungeo}: {stock}개")
```

[문제 5: 관리자 모드]

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
# 초기 붕어빵 재고 설정
bungeo_stock = {
    '팥붕어빵': 10,
    '슈크림붕어빵': 8,
    '초코붕어빵': 5
}

# 관리자 모드 함수
def admin_mode():
    while True:
        print("\n=== 관리자 모드 ===")
        print("1. 붕어빵 재고 추가")
        print("2. 종료")
        choice = input("선택하세요 (1 또는 2): ")
        
        if choice == '1':
            bungeo_type = input("추가할 붕어빵 종류를 입력하세요 (ex. '팥붕어빵'): ")
            if bungeo_type in bungeo_stock:
                try:
                    additional_stock = int(input("추가할 재고 수량을 입력하세요: "))
                    bungeo_stock[bungeo_type] += additional_stock
                    print(f"{bungeo_type} {additional_stock}개 추가 완료")
                    print("\n업데이트된 붕어빵 재고:")
                    for bungeo, stock in bungeo_stock.items():
                        print(f"{bungeo}: {stock}개")
                except ValueError:
                    print("잘못된 입력입니다. 수량은 숫자로 입력하세요.")
            else:
                print("해당 붕어빵은 판매하지 않습니다.")
        
        elif choice == '2':
            print("관리자 모드를 종료합니다.")
            break
        
        else:
            print("잘못된 입력입니다. 다시 선택하세요.")

# 관리자 모드 실행
admin_mode()
```

[문제 6: 매출 확인]

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
# 초기 설정
bungeo_price = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}

bungeo_sold = {
    "팥붕어빵": 5,
    "슈크림붕어빵": 8,
    "초코붕어빵": 3
}

# 총 매출 계산
total_revenue = 0
for bungeo, quantity in bungeo_sold.items():
    price = bungeo_price[bungeo]
    total_revenue += price * quantity

print(f"총 매출: {total_revenue} 원")
```
