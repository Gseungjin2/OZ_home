```python
import random

# 게임 초기화
def initialize_game(n):
    # 보드 크기 설정
    board_size = n
    # 보물 위치 무작위 설정
    treasure_position = (random.randint(0, board_size-1), random.randint(0, board_size-1))
    # 플레이어 위치 무작위 설정
    player_position = (random.randint(0, board_size-1), random.randint(0, board_size-1))
    
    # 보물과 플레이어 위치가 동일하면 다시 설정
    while treasure_position == player_position:
        player_position = (random.randint(0, board_size-1), random.randint(0, board_size-1))
    
    return board_size, treasure_position, player_position

# 거리 계산
def calculate_distance(treasure_position, player_position):
    return abs(treasure_position[0] - player_position[0]) + abs(treasure_position[1] - player_position[1])

# 플레이어 이동
def move_player(board_size, player_position, direction):
    x, y = player_position
    if direction == 'N' and x > 0:
        x -= 1
    elif direction == 'S' and x < board_size - 1:
        x += 1
    elif direction == 'E' and y < board_size - 1:
        y += 1
    elif direction == 'W' and y > 0:
        y -= 1
    return (x, y)

# 게임 실행
def play_game(board_size):
    board_size, treasure_position, player_position = initialize_game(board_size)
    moves = 0

    while True:
        print(f"현재 위치: {player_position}")
        distance = calculate_distance(treasure_position, player_position)
        print(f"보물까지의 거리: {distance}")

        if distance == 0:
            print(f"축하해! 보물을 찾았어! 총 이동 횟수: {moves}")
            break

        direction = input("이동 방향을 입력하세요 (N, S, E, W): ").strip().upper()
        player_position = move_player(board_size, player_position, direction)
        moves += 1

# 게임 보드 크기 설정 및 게임 시작
if __name__ == "__main__":
    board_size = 5  # 보드 크기를 5x5로 설정
    play_game(board_size)
```
