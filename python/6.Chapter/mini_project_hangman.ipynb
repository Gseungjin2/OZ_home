```python
# Í∏∞Î≥∏ ÎùºÏù¥Î∏åÎü¨Î¶¨Îùº ÌååÏù¥Ïç¨ ÏÑ§ÏπòÏãú ÏûêÎèôÏúºÎ°ú ÏÑ§ÏπòÎê©ÎãàÎã§!
# random Î™®ÎìàÏùÄ ÎûúÎç§Ìïú Ïà´ÏûêÎ•º ÏÉùÏÑ±Ìï† Îïå ÏÇ¨Ïö©Ìï©ÎãàÎã§.
import random
import os
# Í≤åÏûÑÏóê ÏÇ¨Ïö©Îê† Îã®Ïñ¥ Î™©Î°ù
words = ["apple", "banana", "orange", "grape", "lemon"]

# ÌñâÎß® Í∑∏Î¶º
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   / \
     |
    ---""",
]


class HangmanGame:
    def __init__(self, word):
        print("Game Start")
        self.word = word
        self.life = 10

    def display(self, life):
        print("========================\n")
        if life > 6:
            print(hangman_pics[0])
            print("========================\n")
        else:
            print(hangman_pics[6-life])
            print("========================\n")
            
    def play(self):
        emptyWord = ["_"] * len(self.word)
        
        while True:
            
            print(f"ÎßûÏ∂ú Îã®Ïñ¥ = {emptyWord}")
            ## Input
            inSpell = input("ÏûÖÎ†• Í∏ÄÏûê : ")
            
            ## Screen Clear
            os.system("cls")
            
            ## Checking & Fill 
            for spell in range(len(self.word)):
                letter = self.word[spell]         
                if letter == inSpell:
                    emptyWord[spell] = inSpell
            
            ## Success or Fail
            if inSpell in emptyWord:
                print("Îã®Ïñ¥Î•º ÎßûÏ∂îÏÖ®Íµ∞Ïöî, Îçî ÌûòÎÇ¥Î¥êÏöî.")
                print(f"ÎÇ®ÏùÄ ÏãúÎèÑÎäî {self.life}Î≤à")
            else:
                self.life -= 1
                print(f"ÌãÄÎ†∏ÏäµÎãàÎã§. ÎÇ®ÏùÄ ÏãúÎèÑÎäî {self.life}Î≤à")
                    
            self.display(self.life)     

            ## Done!
            if self.word == ''.join(emptyWord):
                return print(f"Ï∂ïÌïòÌï©ÎãàÎã§! Îã®Ïñ¥Î•º ÎßûÏ∂îÏÖ®ÏäµÎãàÎã§! \nÎÇ®ÏùÄ ÏãúÎèÑÎäî {self.life}\n")
            
            ## GoodBye!
            if self.life == 0:
                return print("Ï£ΩÏóàÏäµÎãàÎã§! ÏûòÍ∞ÄÏÑ∏Ïöî!") 
            

if __name__ == "__main__":
    game = HangmanGame(random.choice(words))
    game.play()
```
